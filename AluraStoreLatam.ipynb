### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

# 2. Ventas por categoría

```python
ventas1=tienda.groupby('Categoría del Producto').size().sort_values(ascending=False)
print("Cantidad de productos vendidos de la tienda 1, empezando por la más popular son:\n\t ", ventas1)
```

```python
ventass2=tienda2.groupby('Categoría del Producto').size().sort_values(ascending=False)
print("Cantidad de productos vendidos de la tienda 2, empezando por la más popular son:\n\t ", ventass2)
```

```python
ventass3=tienda3.groupby('Categoría del Producto').size().sort_values(ascending=False)
print("Cantidad de productos vendidos de la tienda 3, empezando por la más popular son:\n\t ", ventass3)
```

```python
ventass4=tienda4.groupby('Categoría del Producto').size().sort_values(ascending=False)
print("Cantidad de productos vendidos de la tienda 4, empezando por la más popular son:\n\t ", ventass4)
```

```python
import matplotlib.pyplot as plt

# Asumiendo que ya tienes definidas: ventas_cat1, ventas_cat2, ventas_cat3, ventas_cat4
ventas_list = [ventas1, ventass2, ventass3, ventass4]

# Crear subplots (2 filas, 2 columnas)
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
axes = axes.flatten()  # Convertimos la matriz de ejes en una lista plana para iterar fácilmente

for i in range(4):
    # Graficar en el eje correspondiente
    ventas_list[i].plot(kind="bar", ax=axes[i], color='skyblue', edgecolor='black')
    
    # Configuración de títulos y etiquetas
    axes[i].set_title(f"Ventas por categoría - {Todas_tiendas[i]}", fontsize=12, fontweight='bold')
    axes[i].set_ylabel("Cantidad de ventas")
    axes[i].set_xlabel("Categoría")
    
    # Corregir rotación y alineación de las etiquetas del eje X
    axes[i].set_xticklabels(axes[i].get_xticklabels(), rotation=35, ha='right')
    
    # Añadir cuadrícula ligera para mejor lectura
    axes[i].grid(axis='y', linestyle='--', alpha=0.7)

# Ajustar el diseño para que no se encimen los elementos
plt.tight_layout()
plt.show()
```

# 3. Calificación promedio de la tienda


```python

promedio_t1 = tienda['Calificación'].mean()
promedio_t2 = tienda2['Calificación'].mean()
promedio_t3 = tienda3['Calificación'].mean()
promedio_t4 = tienda4['Calificación'].mean()

print("Promedio de calificaciones de la tienda 1:", promedio_t1)
print("Promedio de calificaciones de la tienda 2:", promedio_t2)
print("Promedio de calificaciones de la tienda 3:", promedio_t3)
print("Promedio de calificaciones de la tiendaa 4:", promedio_t4)
```

```python
promedios = [promedio_t1, promedio_t2, promedio_t3, promedio_t4]
```

```python
plt.figure()
plt.bar(Todas_tiendas, promedios)
plt.title('Promedio de Calificaciones por Tienda')
plt.xlabel('Tiendas')
plt.ylabel('Calificación Promedio')

# Mostrar valores encima de cada barra
for i in range(len(promedios)):
    plt.text(i, promedios[i], f'{promedios[i]:.2f}', ha='center', va='bottom')

plt.show()
```

# 4. Productos más y menos vendidos

```python
ventas1 = tienda.groupby('Producto').size().sort_values(ascending=False)

print("Productos más vendidos en Tienda 1:")
print(ventas1.head())

print("\nProductos menos vendidos en Tienda 1:")
print(ventas1.tail())
```

```python
ventas3= tienda3.groupby('Producto').size().sort_values(ascending=False)

print("Productos más vendidos en Tienda 3:")
print(ventas3.head())

print("\nProductos menos vendidos en Tienda 3:")
print(ventas3.tail())
```

```python
ventas4= tienda4.groupby('Producto').size().sort_values(ascending=False)

print("Productos más vendidos en Tienda 4:")
print(ventas4.head())

print("\nProductos menos vendidos en Tienda 4:")
print(ventas4.tail())
```

```python

tops_ventas = [ventas1.head(5), ventas2.head(5), ventas3.head(5), ventas4.head(5)]

# Crear subplots (2 filas, 2 columnas)
fig, axes = plt.subplots(2, 2, figsize=(15, 12))
axes = axes.flatten()
# Colores diferentes para cada tienda
colores = ['#4C72B0', '#55A868', '#C44E52', '#8172B3']

for i in range(4):
    # Graficamos el Top 5 de la tienda actual
    tops_ventas[i].plot(kind="barh", ax=axes[i], color=colores[i], edgecolor='black')
    
    # Invertir el eje Y para que el producto más vendido esté arriba
    axes[i].invert_yaxis()
    
    # Configuración de títulos y etiquetas
    axes[i].set_title(f"Top 5 mejores Productos - {Todas_tiendas[i]}", fontsize=14, fontweight='bold')
    axes[i].set_xlabel("Cantidad Vendida")
    axes[i].set_ylabel("") # Quitamos el label de 'Producto' para que se vea más limpio
    
    # Añadir los valores al final de cada barra
    for index, value in enumerate(tops_ventas[i]):
        axes[i].text(value + 0.5, index, str(value), va='center', fontsize=11)

plt.tight_layout()
plt.show()
```

```python

tops_peores_ventas = [ventas1.tail(5), ventas2.tail(5), ventas3.tail(5), ventas4.tail(5)]

# Crear subplots (2 filas, 2 columnas)
fig, axes = plt.subplots(2, 2, figsize=(15, 12))
axes = axes.flatten()
# Colores diferentes para cada tienda
colores = ['#4C72B0', '#55A868', '#C44E52', '#8172B3']

for i in range(4):
    # Graficamos el Top 5 de la tienda actual
    tops_peores_ventas[i].plot(kind="barh", ax=axes[i], color=colores[i], edgecolor='black')
    
    # Invertir el eje Y para que el producto más vendido esté arriba
    axes[i].invert_yaxis()
    
    # Configuración de títulos y etiquetas
    axes[i].set_title(f"Top 5 peores Productos - {Todas_tiendas[i]}", fontsize=14, fontweight='bold')
    axes[i].set_xlabel("Cantidad Vendida")
    axes[i].set_ylabel("") # Quitamos el label de 'Producto' para que se vea más limpio
    
    # Añadir los valores al final de cada barra
    for index, value in enumerate(tops_peores_ventas[i]):
        axes[i].text(value + 0.5, index, str(value), va='center', fontsize=11)

plt.tight_layout()
plt.show()
```

# Informe


El presente informe tiene como objetivo evaluar el rendimiento de las cuatro tiendas de Alura Store, con el fin de determinar cuál de ellas debería vender el Sr. Juan para enfocarse en un nuevo emprendimiento. 

La decisión se basa en el análisis de 5 factores clave: ingresos totales, productos y categorías más y menos vendidos, calificaciones promedio de clientes y costos de envío.



1. Ingresos totales por tienda

La tienda 1 se posiciona como la más rentable, mientras que Tienda 4 obtiene los ingresos más bajos. Esta métrica es fundamental, ya que refleja la eficiencia en ventas.

2. Distribución de ventas por categoría

Las categorías más populares en todas las  tiendas son los muebles. Tienda 3 y Tienda 4 destacan por tener una mayor concentración de ventas en categorías con alta demanda en muebles y electrodomesticos. 

3. Calificación promedio de los clientes

Las calificaciones de los clientes reflejan su experiencia de compra. Tienda 3 presenta la mejor calificación promedio, seguida por Tienda 2. Tienda 1, a pesar de sus ingresos altos, tiene una calificación más baja, lo que podría indicar una oportunidad de mejora en atención o logística. Tienda 4 tiene una calificación media.

4. Productos más y menos vendidos

Al observar los productos individuales, Tienda 1 y Tienda 2 presentan una clara presencia de productos líderes en ventas. Tienda 4 tiene un bajo volumen tanto en sus productos más vendidos como en los menos vendidos.

5. Costo de envío promedio

Los costos de envío son un componente importante de la percepción del cliente. Tienda 3 mantiene un costo promedio competitivo. Tienda 1, en cambio, posee uno de los costos de envío más altos, lo que impacta directamente la satisfacción del cliente y la conversión de ventas.

Con base en los datos analizados, la tienda que presenta consistentemente el rendimiento más débil es la Tienda 4. Esta tienda tiene:

* Los ingresos más bajos de toda la cadena.

* Una distribución de ventas por categoría poco destacada.
* Calificaciones promedio inferiores en comparación con las otras tiendas.
* Productos con baja rotación y sin un líder claro en ventas.

Por estas razones, se recomienda que el Sr. Juan considere vender la Tienda 4, ya que presenta debilidades en casi todos los indicadores analizados. Esta decisión le permitirá concentrar sus recursos en las tiendas con mayor potencial de crecimiento y rentabilidad.
